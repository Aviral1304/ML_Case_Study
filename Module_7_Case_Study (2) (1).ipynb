```python
# Model Building 1
```

```python
#Separting dependent and independent variable
x=leads_df.drop(['Converted'],axis=1)
y=leads_df['Converted']
```

```python
x.columns
```

```python
#Categorical to Numerical
from sklearn.preprocessing import LabelEncoder
sm=LabelEncoder()
x['Lead_Origin']=sm.fit_transform(x['Lead_Origin'])
x['Lead_Source']=sm.fit_transform(x['Lead_Source'])
x['Do_Not_Email']=sm.fit_transform(x['Do_Not_Email'])
x['TotalVisits']=sm.fit_transform(x['TotalVisits'])
x['Total_Time_Spent_on_Website']=sm.fit_transform(x['Total_Time_Spent_on_Website'])
x['Page_Views_Per_Visit']=sm.fit_transform(x['Page_Views_Per_Visit'])
x['Last_Activity']=sm.fit_transform(x['Last_Activity'])
x['Country']=sm.fit_transform(x['Country'])
x['Specialization']=sm.fit_transform(x['Specialization'])
x['Current_Occupation']=sm.fit_transform(x['Current_Occupation'])
x['Tags']=sm.fit_transform(x['Tags'])
x['City']=sm.fit_transform(x['City'])
x
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
#Splitting data into training and testing dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Logistic model
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0) #random_state=0 is optional
classifier.fit(x_train,y_train)
```

```python
#Predicting Value for test data
y_pred=classifier.predict(x_test)
y_pred
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#Accuracy of model
accuracy_score(y_test,y_pred)
```

```python
#Classification report for the model
from sklearn.metrics import classification_report 
print(classification_report(y_test,y_pred))
```
